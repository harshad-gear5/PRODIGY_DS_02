```python
import pandas as pd

# Load the Titanic dataset
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_df = pd.read_csv(url)

# Display the first few rows of the dataset
titanic_df.head()
```

```python
# Display information about the dataset
titanic_df.info()

# Summary statistics of numerical columns
titanic_df.describe()
```

```python
# Check for missing values
titanic_df.isnull().sum()

# Drop rows with missing values
titanic_df_clean = titanic_df.dropna()

# Verify if there are any missing values left
titanic_df_clean.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Relationship between survival and gender
sns.countplot(x='Survived', hue='Sex', data=titanic_df_clean)
plt.title('Survival Count by Gender')
plt.show()

# Relationship between survival and ticket class
sns.countplot(x='Survived', hue='Pclass', data=titanic_df_clean)
plt.title('Survival Count by Ticket Class')
plt.show()

# Distribution of age among passengers
plt.figure(figsize=(8, 5))
sns.histplot(titanic_df_clean['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Fare distribution by ticket class
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', data=titanic_df_clean)
plt.title('Fare Distribution by Ticket Class')
plt.show()
```

```python
Survival rates were higher among females compared to males.
Passengers in higher ticket classes (1st class) had better survival rates.
The majority of passengers were in the age range of 20-40 years.
Fare generally increased with ticket class, indicating a correlation between class and fare.
```
